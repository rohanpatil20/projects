### Import necessary modules

```python
import car_rental as cr
```

### Set Constants

```python
# Setting constants that will be used for determining the action that the user will select
INPUT_TYPE_VIEW = 1
INPUT_TYPE_RENT = 2
INPUT_TYPE_RETURN = 3
INPUT_TYPE_EXIT = 4
```

```python
def main():
    store = cr.CarRentalStore()
    store.addCar(1, 'Honda', 'Civic', 9.25)
    store.addCar(2, 'Toyota', 'Corolla', 9.50)
    store.addCar(3, 'Volkswagen', 'Tiguan', 12.50)
    store.addCar(4, 'Tesla', 'Model S', 25)
    store.addCar(5, 'Ford', 'F-150', 15)
    
    store.addCustomer(1, 'John Doe', 'johndoe', 'johndoe2141')
    store.addCustomer(2, 'Jane Doe', 'janedoe', 'janedoe1412')
    
    currentUser = None
    while currentUser is None:
        userName = input("Enter username:")
        password = input("Enter password:")
        
        userFound = False
        
        for i in range(len(store.customers)):
            print(store.customers[i].userName)
            if userName == store.customers[i].userName:
                userFound = True
                if store.customers[i].verifyLogin(password):
                    currentUser = store.customers[i]
                    print("Welcome " + store.customers[i].name)
                    break
                else:
                    print("Incorrect Password. Please try again.")
                    break
                    
        if not userFound:
            print("Username not found. Please try again.")
    
    storeInUse = True

    while storeInUse:
        try: 
            input_type = int(input(
                '''
            Select action by entering the number associated with it:
            1. View available cars
            2. Rent a car
            3. Return car
            4. Exit
            '''
            ))
            if input_type == INPUT_TYPE_EXIT:
                storeInUse = False
                continue
            elif input_type == INPUT_TYPE_VIEW:
                store.getAvailableCars()
                continue
            elif input_type == INPUT_TYPE_RENT:
                rent_car_id = 0
                while rent_car_id == 0:
                    try:
                        print('Select a car to rent by entering the number associated with it:')
                        available_cars_string = ''
                        for car in store.cars:
                            if car.is_rented == False:
                                available_cars_string += str(car) + '\n'
                        rent_car_id = int(input(available_cars_string))
                        if rent_car_id <= 0:
                            rent_car_id = 0
                            print("Invalid car. Please try again.")
                            continue
                    except ValueError:
                        rent_car_id = 0
                        print("Invalid car. Please try again.")
                
                selected_rental_method = 0
                while selected_rental_method == 0:
                    try:
                        print('Select rental method by entering the number associated with it:')
                        rental_methods_string = ''
                        for rent_type in store.RENT_TYPES:
                            rental_methods_string += str(store.RENT_TYPES[rent_type]) + '. ' + rent_type + '\n' 
                        selected_rental_method = int(input(rental_methods_string))
                        
                        if selected_rental_method not in store.RENT_TYPES.values():
                            selected_rental_method = 0
                            print("Invalid rental method. Please try again.")
                            continue        
                    except ValueError:
                        selected_rental_method = 0
                        print("Invalid rental method. Please try again.")
                
                rent_time_amount = 0
                time_unit_type = 'hours'
                if selected_rental_method == store.RENT_TYPES['Daily']:
                    time_unit_type = 'days'
                elif selected_rental_method == store.RENT_TYPES['Weekly']:
                    time_unit_type = 'weeks'
                        
                while rent_time_amount == 0:
                    try:
                        rent_time_amount = float(input("Enter how many " + time_unit_type + " to rent the car for:"))
                    
                        if rent_time_amount <= 0:
                            rent_time_amount = 0
                            print("Invalid unit value. Please try again.")
                            continue
                    except ValueError:
                        rent_time_amount = 0
                        print("Invalid unit value. Please try again.")
                
                store.rentCar(rent_car_id, currentUser.id, selected_rental_method, rent_time_amount)
                continue
            elif input_type == INPUT_TYPE_RETURN:
                return_car_id = 0
                return_cars_string = ''
                car_count = 0
                for car in store.cars:
                    if car.is_rented == True and car.renter_id == currentUser.id:
                        return_cars_string += str(car) + '\n'
                        car_count += 1
                        
                if car_count == 0:
                    print('No cars to return.')
                    continue
                        
                while return_car_id == 0:
                    print('Select which car to return by entering the number associated with it: ')
                    try:
                        return_car_id = int(input(return_cars_string))
                        if return_car_id <= 0:
                            return_car_id = 0
                            print("Invalid car. Please try again.")
                            continue
                    except ValueError:
                        return_car_id = 0
                        print("Invalid car. Please try again.")
                        
                store.returnCar(return_car_id, currentUser.id)
                continue
                    
            else:
                input_type = 0
                print("Invalid action. Please try again.")
                continue
        except ValueError:
            input_type = 0
            print("Invalid action. Please try again.")
main()          
```
